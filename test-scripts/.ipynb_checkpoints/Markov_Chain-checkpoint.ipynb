```python
f = open('text.txt', 'r', encoding="utf8")
```

```python
text = f.read()
```

```python
import random
```

```python
a = text.replace('\n', '***')
a = a.replace('\t', ' ')
a = a.replace('\xa0', ' ')
a = a.replace('   ', ' ')
b = a.split(' ')
sec = {}
for i in range(len(b)-2):
    seca = ' '.join([b[i], b[i+1]])
    if seca not in sec:
        sec[seca] = [b[i+2]]
    else:
        sec[seca].append(b[i+2])
```

```python
b
```

```python
text2 = random.choice(list(sec.keys()))
data2 = text2.split()
for i in range(5000):
    seca = ' '.join([data2[i], data2[i+1]])
    data2.append(random.choice(sec[seca]))
pri = ' '.join(data2)
asa = pri.split('******')
for i in asa:
    print(i)
```
